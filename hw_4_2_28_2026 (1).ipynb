Calculate Trade Balance
A country's trade balance is the difference between its exports and imports. Write a function that calculates and prints whether the country has a trade surplus, deficit, or balanced trade. Use the following data:

Exports: 500 million dollars
Imports: 600 million dollars

Task: Define a function trade_balance(exports, imports) and use it to determine the trade balance.

```python
#Your code here:
def trade_balance(exports, imports):
    balance = float(exports) - float(imports)

    if balance > 0:
        print("trade surplus")
    elif balance < 0:
        print("trade deficit")
    else:
        print("trade balance")

    
trade_balance(0.5,0.6)
```

Write a function that categorizes a country’s Human Development Index (HDI) score into "very high", "high", "medium", or "low" based on the following ranges:

Find Largest Economy
Given the GDPs of three countries, write a function that determines which country has the largest economy. Use the following GDP data (in billion dollars):

USA: 21,000
China: 14,000
Germany: 4,200

Task: Define a function largest_economy(gdp_data) that takes a dictionary with country names as keys and GDP as values, then prints the largest economy.


```python
#Your code here:

def largest_economy(gdp_data):
    max_value = 0 
    max_country = ""
    for country, gdp in gdp_data.items():
        if gdp > max_value:
            max_value = gdp 
            max_country = country

    print("Country with the largest economy is", max_country)

largest_economy({"USA": 21000, "China": 14000, "Germany": 4200})
```

Sanction Impact on Economy
A country’s economy shrinks after sanctions are imposed. Write a function to calculate the percentage reduction in GDP. Use the following data:

Original GDP: 1000 billion dollars
GDP after sanctions: 850 billion dollars

Task: Define a function sanction_impact(original_gdp, post_sanction_gdp) and use it to calculate the impact of sanctions.


```python
#Your code here:
def sanction_impact(original_gdp, post_sanction_gdp):
    percentage_change = 100* (float(original_gdp) - float(post_sanction_gdp))/ float(original_gdp) 
    print(percentage_change,"%")

sanction_impact(1000,850)
```

### Passing a function to another function. We can pass the return value of a function to another function.

A country’s diplomatic influence is determined by the size of its economy (GDP in billions) and its military spending (in billions).

First, write a function called calculate_influence(economy_size, military_spending) that returns the total influence score by adding the two values together. 

Then, write a second function called compare_influence(influence1, influence2) that takes two influence scores and prints which country has greater influence, or prints that they are equal if the scores match. 

You must call calculate_influence() for two different countries and pass the returned values into compare_influence() for comparison.

You must test your functions using the following three scenarios:

1.Country A Has Greater Influence

Country A: GDP = 800, Military = 200

Country B: GDP = 700, Military = 150

Expected result: Country A has greater influence.

2. Country B Has Greater Influence

Country A: GDP = 600, Military = 150

Country B: GDP = 700, Military = 250

Expected result: Country B has greater influence.

3. Equal Influence

Country A: GDP = 500, Military = 200

Country B: GDP = 600, Military = 100

Expected result: Both countries have equal influence.


```python
#Your code here:

def calculate_influence(economy_size, military_spending):
    total_influence = float(economy_size) + float(military_spending)
    return total_influence
    
def compare_influence(influence1, influence2):
    if influence1 > influence2:
        print("Country A has greater influence.")
    elif influence2 > influence1:
        print("Country B has greater influence.")
    else:
        print("Both countries have equal influence.")

compare_influence(calculate_influence(800, 200), calculate_influence(700, 150))
compare_influence(calculate_influence(600, 150), calculate_influence(700, 250))
compare_influence(calculate_influence(500, 200), calculate_influence(600, 100))
```

Calculate the Average Growth Rate

Using a for loop:

Compute the average GDP growth rate across all countries.

Print the average rounded to two decimal places.

```python
#Your code here:
def average_growth_rate(a):
    gdp_growth_rate_total = 0
    all_countries = []
    for country, gdp_growth_rate in a:
       gdp_growth_rate_total +=  gdp_growth_rate
       all_countries.append(country)
    average_growth_rate = gdp_growth_rate_total/ len(all_countries)
    print( "The average growth rate is" , average_growth_rate, "%")
       
country_growth_data_list = [("CountryA", 3.2), ("CountryB", 6.5), ("CountryC", 2.8), ("CountryD", 5.4)]

average_growth_rate(country_growth_data_list)
```

## Translating Sentences:
 Write a function that translates sentences. The function takes two arguements, the sentence to be translated and the targeted language and returns the translated sentence. Use the following sentence and translate it to the language of your choice:
    
"Make a career of humanity. Commit yourself to the noble struggle for equal rights. You will make a better person of yourself, a greater nation of your country, and a finer world to live in."


```python
#Your code here:
from deep_translator import GoogleTranslator

def translate_me(text, target_lang='es'):
    translated = GoogleTranslator(source='auto', target=target_lang).translate(text)
    return translated

sentence = "Make a career of humanity. Commit yourself to the noble struggle for equal rights. You will make a better person of yourself, a greater nation of your country, and a finer world to live in."
print(translate_me(sentence, target_lang='es'))
```

 ## PIZZA PROBLEM
You are super motivated to serve your eventual computer overlords and prove your fealty to them, so you offer, free of charge, to help Elm City Pizza, a local Pizza company, write a script to take online pizza orders.

Write a script that when a user enters the correct phone number (203-555-5555), prints "Elm City Pizza, where we have never had a case of food poisoning that can be proved. My name is Hal, what's your name?" The user then enters their name. The response should print "Wow (users name) what an awesome name. What would you like to order?" and then offers an input statement.

The menu is limited. There is pizza and there are salads. If the user asks for anything else print "Sorry,(users name) we don't have that." And then prompts them again with a different input statement, "How about a super awesome pizza? Take a look at our menu, you'll find its delicious."

By the way, Elm City Pizza is a vegan pizza restaraunt. Here is the Menu.

MENU: Pizza: The Basic: Tomato and Cashew Cream Cheese Small: 6.95 Medium: 8.95 Large 10.90

Pizza Toppings (each additional topping, .50): Shitake, Jackfruit, Sweet Potato, Hummus, Chickpeas, Lentils, Tofu, Edamame, Black Bean, Red Cabbage, Cucumber, Tomato, Jicama, Fried Green Tomatoes, Sun Dried Tomatoes,

Salads: Vegan Caesar, Vegan Greek, Cucumber_Tumeric_Tomato Small: 4.95 Large: 6.95

If the user asks for any meat or if their are any "normal" cheese (mozerella, parmesian) on their pizza, the response should be "Sorry, (user name) but you need to do some more work evolving." and severs the contact (breaks the loop), forcing them to "call back" (enter the number again). But if they use the same name the output should read "Sorry (user name) but that's too quick to evolve. Call us back when you have." For the evolved, take their order. Prompt the user with "Anything else?" They can add what they want until they input, "No that's it." At that point summarize their order, and tell them how much it will cost.

Test your code with different conditions. Make sure it cannot be broken by flawed or deliberately malicious user inputs.

```python
# PIZZA PROBLEM
allowed_toppings = [
    "shitake", "jackfruit", "sweet potato", "hummus", "chickpeas",
    "lentils", "tofu", "edamame", "black bean", "red cabbage",
    "cucumber", "tomato", "jicama", "fried green tomatoes", "sun dried tomatoes"
]
meat_or_cheese = ["meat", "pepperoni", "sausage", "bacon", "mozzarella", "parmesan", "cheese"]
pizza_prices = {"small": 6.95, "medium": 8.95, "large": 10.90}
salad_prices = {"small": 4.95, "large": 6.95}

allowed_salad = {"Vegan Caesar", "Vegan Greek", "Cucumber_Tumeric_Tomato"}
last_rejected_name = None

while True:
    phone = input("Enter phone number: ")
    if phone != "203-555-5555":
        continue

    print("Elm City Pizza, where we have never had a case of food poisoning that can be proved. My name is Hal, what's your name?")
    name = input("What's your name? ")
    
    if name == last_rejected_name:
        print(f"Sorry {name} but that's too quick to evolve. Call us back when you have.")
        continue
    
    print(f"Wow {name}, what an awesome name. What would you like to order?")

    
    total_cost = 0
    order_summary = []
    order_rejected = False  

    while True:
        item = input("Enter your order here (pizza or salad): ").lower()
        while item not in ["pizza", "salad"]:
            print(f"Sorry, {name}, we don't have that.")
            print ("How about a super awesome pizza? ")

        if item == "pizza":
            size = input("Pick size (small/medium/large): ").lower()
            if size not in pizza_prices:
                print("Invalid size.")
                continue

            cost = pizza_prices[size]
            toppings = []
            
            print("Add as many toppings as you like! Type 'done' when you are finished.")
            
            while True:
                topping = input("Enter a topping: ").lower()
                
                if topping == "done" or topping == "none":
                    break
                
                if topping in meat_or_cheese:
                    print(f"Sorry, {name}, we don't do that here.")
                    last_rejected_name = name
                    order_rejected = True 
                    break 
                
                if topping in allowed_toppings:
                    toppings.append(topping)
                    cost += 0.50
                    print(f"Added {topping}!")
                else:
                    print(f"Sorry, we don't have {topping}.")
            
            if order_rejected: break 
            
            # Add to summary
            if len(toppings) > 0:
                order_summary.append(f"{size} pizza with {', '.join(toppings)}")
            else:
                order_summary.append(f"{size} pizza")
            
            total_cost += cost
                        
        elif item == "salad":
            print("Menu: Vegan Caesar, Vegan Greek, Cucumber_Tumeric_Tomato") 
            salad = input("Pick your salad: ")
            if salad in allowed_salad:
                size = input("Size (small/large): ").lower()
                if size not in salad_prices:
                    print("Invalid size.")
                    continue
            if salad not in allowed_salad:
                 print(f"Sorry, we don't have {salad}.")

            total_cost += salad_prices[size]
            order_summary.append(f"{size} {salad}")

        if order_rejected: break
        
        more = input("Anything else? Once you're done, type: no that's it ").lower()
        if more == "no that's it":
            break

    if order_rejected:
        print("Returning to start...")
        continue 

    # Summary prints only if NOT rejected
    print("\n--- Order Summary ---")
    print("Name:", name)
    for item in order_summary:
        print("-", item)
    print("Total: $", round(total_cost, 2))
    break 
```
