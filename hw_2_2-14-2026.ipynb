## What's a Tuple?

A tuple in Python is an immutable, ordered collection of elements, similar to a list. However, unlike lists, tuples cannot be modified after they are created, meaning you cannot add, remove, or change elements once the tuple is defined.

Key Characteristics of Tuples:

Ordered: The elements in a tuple are ordered, and you can access them using an index (just like in lists).

Immutable: Once a tuple is created, its contents cannot be changed (i.e., no adding, removing, or changing elements). But as we saw above, a tuple can be added to a list. 

Can contain mixed data types: A tuple can hold elements of different types, such as integers, strings, floats, and even other tuples or lists.

Indexable and iterable: You can access elements using an index and iterate over the elements using a loop.

Allows duplicates: Tuples can contain duplicate elements, just like lists

### Creating a Tuple:
Tuples are created by enclosing the elements inside parentheses () and separating them by commas.

```python
# Creating a tuple with mixed data types
my_tuple = (1, 'apple', 3.14, 'banana')

print(my_tuple)       # Output: (1, 'apple', 3.14, 'banana')
print(my_tuple[1])    # Output: 'apple'
```

No, you cannot do new_variable = new_numbers.reverse().

If you try this, new_variable will be equal to None.

Why this happens
In Python, the .reverse() method is an "in-place" operation. This means it modifies the existing list and doesn't "return" a result to the variable. Python purposely returns None here as a safety feature to let you know that you have mutated the original list instead of creating a new one.

How to actually create a new variable
If you want to keep new_numbers as it is and save a reversed version into a new variable, use one of these two "Pythonic" ways:

Option 1: The "Slicing" Way (Most Common)
Using [::-1] tells Python to "start at the end and step backward by 1." It creates a brand new list.

Python
reversed_list = new_numbers[::-1]
Option 2: The reversed() Function
The reversed() function creates an iterator, so you have to wrap it in list() to turn it back into a list you can see.

Python
reversed_list = list(reversed(new_numbers))

In Python, these three tools are used for different "deletion" scenarios. Here is how they break down:

1. del  Delete by Index (or Slice)
del is a keyword, not a method. It is best used when you know exactly where an item sits in the list. It can also delete a range of items at once.

Single Item: del list_name[index]

Range (Slice): del list_name[1:3] (removes items at index 1 and 2)

Python
composers = ["Bach", "Mozart", "Saz", "Beethoven"]
del composers[1] 
# Result: ["Bach", "Saz", "Beethoven"]
2. pop()  Remove and "Capture"
pop() is a method that removes an item and returns it. This is useful if you want to delete something from a list but save it into a variable to use later.

Last Item: list_name.pop() (if you leave the parentheses empty, it defaults to the last item).

Specific Item: list_name.pop(1) (removes item at index 1).

Python
composers = ["Bach", "Mozart", "Saz"]
last_one = composers.pop() 

print(last_one)  # Output: "Saz"
print(composers) # Output: ["Bach", "Mozart"]
3. remove()  Delete by Item Name (Value)
This is the one you asked about specifically. It searches the list for the first occurrence of a specific value and deletes it. You don't need to know the index; you just need to know the name.

Syntax: list_name.remove("value")

Python
composers = ["Bach", "Mozart", "Saz", "Mozart"]
composers.remove("Mozart")

# Result: ["Bach", "Saz", "Mozart"] 
# Note: It only removed the FIRST "Mozart" it found.
