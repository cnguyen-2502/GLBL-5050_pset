Disaster Relief Resource Tracking System

An international humanitarian organization is developing a simple Python program to track incoming donations and the distribution of relief goods during a disaster response. The system must manage donations from different types of donors and allow multiple relief teams to report their distribution progress.

First, define a class called ReliefTeam. Each relief team should have a name and keep track of the quantity of five categories of goods distributed: sanitary supplies, food, clothing, homeware, and entertainment items. All distribution values should initially be zero. The class should include a method report_distribution(category, amount) that increases the amount of goods distributed in the specified category, and a method show_progress() that prints a summary of the team’s distribution progress.

Next, define a class called Donor with attributes for the donor’s name, donor type (which may be "country", "individual", or "organization"), and the amount donated. Include a method that prints a formatted description of the donation.

The program should maintain a dictionary called relief_teams that stores three teams: "A", "B", and "C", each associated with a ReliefTeam object. It should also maintain a dictionary called donations that records donor names and the amount they contributed.

Write a function add_donation() that asks the user to enter a donor name, donor type, and donation amount. The function should create a Donor object and record the donation in the donations dictionary.

Write a function team_login() that allows a relief team to log into the system by selecting team "A", "B", or "C". After logging in, the team should see a menu that allows them to report the distribution of goods by entering a category and amount, view their current distribution progress, or log out.

Write another function admin_portal() that allows an administrator to monitor the system. The administrator should be able to view the total amount of donations received, list all donors and their contributions, and display the distribution progress for all relief teams.

Finally, write a main program loop that repeatedly displays the following menu until the user chooses to exit: donate to the relief effort, log in as a relief team, access the administrator portal, or exit the program. The program should call the appropriate functions based on the user’s selection.

```python
class ReliefTeam:
    def __init__(self, name):
        self.name = name
        self.distributed = {
            "sanitary supplies": 0,
            "food": 0,
            "clothing": 0,
            "homeware": 0,
            "entertainment items": 0
        }

    def report_distribution(self, category, amount):
        if category in self.distributed:
            self.distributed[category] += amount
            print(f"Updated: {category} +{amount} units recorded for Team {self.name}.")
        else:
            print(f"Invalid category '{category}'. Please choose from: {', '.join(self.distributed.keys())}.")

    def show_progress(self):
        print(f"\n  Distribution Progress — Team {self.name}:")
        for category, amount in self.distributed.items():
            print(f"  {category.capitalize():<25} {amount:>6} units")


class Donor:
    VALID_TYPES = ["country", "individual", "organization"]

    def __init__(self, name, donor_type, amount):
        self.name = name
        self.donor_type = donor_type
        self.amount = amount

    def describe_donation(self):
        print(f"\n  Donor:  {self.name}")
        print(f"Type:   {self.donor_type.capitalize()}")
        print(f"Amount: ${self.amount:,.2f}")

relief_teams = {
    "A": ReliefTeam("A"),
    "B": ReliefTeam("B"),
    "C": ReliefTeam("C"),
}

donations = {}

def add_donation():
    print("\nDonation Portal:")
    name = input("Enter donor name: ")
    if not name:
        print("Donor name cannot be empty.")
        return

    donor_type = input("Enter donor type (country / individual / organization): ").strip().lower()
    if donor_type not in Donor.VALID_TYPES:
        print(f"  Invalid donor type. Must be one of: {', '.join(Donor.VALID_TYPES)}.")
        return

    try:
        amount = float(input("Enter donation amount ($): ").strip())
        if amount <= 0:
            raise ValueError
    except ValueError:
        print("Invalid amount. Please enter a positive number.")
        return

    donor = Donor(name, donor_type, amount)
    donor.describe_donation()

    if name in donations:
        donations[name] += amount
        print(f"Existing donor — updated total: ${donations[name]:,.2f}")
    else:
        donations[name] = amount
        print("Donation recorded successfully. Thank you!")


def team_login():
    print("\nRelief Team Login:")
    team_id = input("Enter your team (A / B / C): ").upper()

    if team_id not in relief_teams:
        print("Invalid team. Please choose A, B, or C.")
        return
    
    if team_id == "A":
        password = input("Input password:")
        if password != "glbl5050a":
            print("Invalid password")
            return
  
    if team_id == "B":
        password = input("Input password:")
        if password != "glbl5050b":
            print("Invalid password")
            return
    
    if team_id == "C":
        password = input("Input password:")
        if password != "glbl5050c":
            print("Invalid password")
            return

    team = relief_teams[team_id]
    print(f"Welcome, Team {team_id}!")
    
    CATEGORIES = list(team.distributed.keys())

    while True:
        print(f"\n  Team {team_id} Menu")
        print("1. Report distribution")
        print("2. View progress")
        print("3. Log out")
        choice = input("\n  Select an option: ")

        if choice == "1":
            print(f"\nCategories: {', '.join(CATEGORIES)}")
            category = input("Enter category: ").lower()
            try:
                amount = int(input("  Enter quantity distributed: ").strip())
                if amount <= 0:
                    raise ValueError
            except ValueError:
                print("  Invalid quantity. Please enter a positive whole number.")
                continue
            team.report_distribution(category, amount)

        elif choice == "2":
            team.show_progress()

        elif choice == "3":
            print(f"Team {team_id} logged out.")
            break

        else:
            print("Invalid option. Please select 1, 2, or 3.")


def admin_portal():
    print("\n--- Administrator Portal ---")

    while True:
        print("\n  Admin Menu")
        print("1. View total donations")
        print("2. List all donors")
        print("3. View all teams' distribution progress")
        print("4. Return to main menu")
        choice = input("\n  Select an option: ").strip()

        if choice == "1":
            total = sum(donations.values())
            print(f"\n  Total donations received: ${total:,.2f}")

        elif choice == "2":
            if not donations:
                print("\n  No donations recorded yet.")
            else:
                print(f"\n  {'Donor':<30} {'Amount':>12}")
                for name, amount in donations.items():
                    print(f"  {name:<30} ${amount:>10,.2f}")
                print(f"  {'TOTAL':<30} ${sum(donations.values()):>10,.2f}")

        elif choice == "3":
            for team in relief_teams.values():
                team.show_progress()

        elif choice == "4":
            print("Returning to main menu.")
            break

        else:
            print("Invalid option. Please select 1–4.")


def main():
    print("Disaster Relief Resource Tracking System:")

    while True:
        print("\nMain Menu:")
        print("1. Donate to the relief effort")
        print("2. Log in as a relief team")
        print("3. Access the administrator portal")
        print("4. Exit")

        choice = input("\nSelect an option: ").strip()

        if choice == "1":
            add_donation()
        elif choice == "2":
            team_login()
        elif choice == "3":
            admin_portal()
        elif choice == "4":
            print("\nThank you for supporting disaster relief. Goodbye!\n")
        else:
            print("Invalid option. Please select 1, 2, 3, or 4.")


if __name__ == "__main__":
    main()
```
